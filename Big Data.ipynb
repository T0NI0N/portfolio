Questa immagine rappresenta un foglio di excel dove vi sono due datset dove per ognuno viene calcolata la media dei valori, la distribuzione normale e della deviazione standard.

Questo esempio serve a dimostrare che nonostante le due medie dei dataset siano praticamente uguali, il loro andamento sia totalmente diverso e grazie alla distribuzione normale e alla deviazione standard lo mostriamo nei grafici.

```python
import math

########################### Sviluppo del task in versione python #############################################

#definisco il dataset del produttore a
valProdA = [85, 85, 80, 70, 70, 60, 55, 50, 40, 40, 30, 30, 20, 20, 10]
#stampo il i valori del dataset a
print(f"Dati produttore A: {valProdA}")

#effettuo il calcolo della media, la varianza e della deviazione standard
mean = sum(valProdA) / len(valProdA)
variance = sum((v-mean)**2 for v in valProdA) / len(valProdA)
stdev = math.sqrt(variance)

#stampo i risultati ottenuti dai calcoli
print("Media produttore A: {:.2f}".format(mean))
print("Varianza produttore A: {:.2f}".format(variance))
print("Deviazione standard produttore A: {:.2f}".format(stdev))

########################################################################
print(" ")
print("---------------------------------------------------------------------------------")
print(" ")

#definisco il datset del produttore b
valProdB = [49, 50, 50, 50, 50, 50, 50, 50, 50, 50, 50, 50, 50, 50, 50]
#stampo i valori del dataset b
print(f"Dati produttore B: {valProdB}")

#effettuo il calcolo della media, la varianza e della deviazione standard
mean = sum(valProdB) / len(valProdB)
variance = sum((v-mean)**2 for v in valProdB) / len(valProdB)
stdev = math.sqrt(variance)

#stampo i risultati ottenuti dai calcoli
print("Media produttore B: {:.2f}".format(mean))
print("Varianza produttore B: {:.2f}".format(variance))
print("Deviazione standard produttore B: {:.2f}".format(stdev))
```

```python
###############   TASK 1    ##################

#importo le librerie che servono per richiamare i metodi
import random
import matplotlib.pyplot as plt

# numero delle facce del dado
NUM_SIDES = 6

# array del dado contenente i numeri delle faccie del dado
dado = [x for x in range(1, NUM_SIDES+1)]
sums = []
da1 = []
da2 = []

# itero i dadi per il calcolo delle somme
for i in dado:
    for j in dado:
        tot = i + j
        da1.append(i)
        da2.append(j)
        sums.append(tot)

# calcolo della frequenza di ogni somma
freq = [sums.count(i) for i in range(2, 2*len(dado)+1)]

# calcolo della probabilità
prob = [f/len(sums) for f in freq]


p2 = []
p3 = []
p4 = []
p5 = []
p6 = []
p7 = []
p8 = []
p9 = []
p10 = []
p11 = []
p12 = []

for i in range(len(da1)):
  if da1[i]+da2[i] == 2:
    p2.append((da1[i],da2[i]))
  if da1[i]+da2[i] == 3:
    p3.append((da1[i],da2[i]))
  if da1[i]+da2[i] == 4:
    p4.append((da1[i],da2[i]))
  if da1[i]+da2[i] == 5:
    p5.append((da1[i],da2[i]))
  if da1[i]+da2[i] == 6:
    p6.append((da1[i],da2[i]))
  if da1[i]+da2[i] == 7:
    p7.append((da1[i],da2[i]))
  if da1[i]+da2[i] == 8:
    p8.append((da1[i],da2[i]))
  if da1[i]+da2[i] == 9:
    p9.append((da1[i],da2[i]))
  if da1[i]+da2[i] == 10:
    p10.append((da1[i],da2[i]))
  if da1[i]+da2[i] == 11:
    p11.append((da1[i],da2[i]))
  if da1[i]+da2[i] == 12:
    p12.append((da1[i],da2[i]))


print("Tutte le possibili combinazioni sono:")
print("2  = {",p2,"}")
print("3  = {",p3,"}")
print("4  = {",p4,"}")
print("5  = {",p5,"}")
print("6  = {",p6,"}")
print("7  = {",p7,"}")
print("8  = {",p8,"}")
print("9  = {",p9,"}")
print("10 = {",p10,"}")
print("11 = {",p11,"}")
print("12 = {",p12,"}")
print("Possibilità totali: P(A intersecato B)= n/", len(sums))
print("Frequenza: ", freq)

# creazione grafico
plt.step(range(2, 2*len(dado)+1), prob)
plt.title("PDF")
plt.xlabel("Possibili risultati")
plt.ylabel("Probabilità")
plt.show()
```

```python
################## TASK 2 - Pearson vs. Spearman ######################

#importo le librerie che serviranno per richiamre i metodi
import matplotlib.pyplot as plt
import numpy as nm
import scipy.stats as sci

#Creo i dataset della tenperatura e del Vi
T = [0,5,10,15,20,30,40,60,80,100,150,200,250,300,374]
Vi = [0.0179,0.0179,0.0179,0.0179,0.0179,0.0180,0.0180,0.0183,0.0185,0.0187,0.0196,0.2070,0.0224,0.0252,0.0570]

#calcolo il coefficiente di pearson tra i due dataset
pearsonCoefficient = sci.pearsonr(T,Vi)

#calcolo il rango dei due dataset
rankT = sci.rankdata(T,'average')
rankVi = sci.rankdata(Vi,'average')

#calcolo il coefficiente di spearman tra i valori dei rank sopra calcolati
spearmanCoefficient = sci.spearmanr(rankT,rankVi,0,'propagate','two-sided')

#creo il grafico
plotter = plt.subplot()

plotter.scatter(T,Vi, c='yellow', edgecolors='black')
z=nm.polyfit(T,Vi,1)
p=nm.poly1d(z)

plotter.plot(T,p(T),color="purple", linewidth=3, linestyle="dotted")

plt.title('VI (I)')
plt.xlabel('T')
plt.ylabel('Vi')

plt.show()

#stampo i valori dei coeffienti di pearson e di spearman
print("Il valore del coefficiente di pearson è: ",round(pearsonCoefficient[0],3))
print("Il valore del coefficiente di spearman è: ",round(spearmanCoefficient[0],3))
```

In questo task dati due dataset, abbiamo calcolato il coeffiente di spearman per determinare la correlazione tra ranghi ( per cui è stato effettuato il calcolo del rango) e il coefficiente di pearson per determinare se i due dataset sono correlati linearmente ( il valore varia da -1 a 1 dove: 1 è correlato positivamente, 0 invece corrisponde a fatto che i due dataset non sono correlati affatto , mentre -1 indica che vi è una correlazione negativa).

```python
################## Task3: Correlazione e Periodicità ##################

#importo le librerie che serviranno per richiamre i metodi
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

#inizializzo lo shift che verrà appricato
SHIFT = 7

#creo il dataset dei dati
data = [100, 90, 80, 80, 90, 60, 40, 105, 95, 80, 85, 95, 60, 35, 100, 90, 75, 80, 90, 65, 40, 105, 90, 80, 85, 90, 55, 35]
#creo il dataset del leg+7
values = [x for x in range(len(data))]

#effettuo lo shift di 7 posizioni per i 7 giorni e calcolo il valore del coefficiente di pearson
base = data[0:7]
y = []

for i in range(1, len(data) + 1 - SHIFT):
	shifted = data[(i):(i + SHIFT)]
	corr, _ = pearsonr(base, shifted)
	y.append(corr)


#creo il grafico per mostrare i risuktati
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(9, 4))

ax1.plot(values, data, 'r')
ax1.set_title("Traffico a Milano")
ax1.set_xlabel("Giorno")
ax1.set_ylabel("Indice traffico")

ax2.plot([x for x in range(len(y))], y, 'o-')
ax2.axhline(0, color='blue', linestyle=':')
ax2.set_title("Autocorrelazione")
ax2.set_xlabel("Shift")
ax2.set_ylabel("Coeff. Autocorr.")

fig.tight_layout()
plt.show()
```

Spesso i dati presentano ciclicità (o pattern), notiamo infatti che l'andamento del primo grafico tende a ripetersi, per confermare e identificare la ciclicità, nel secondo grafico viene correlata la sequenza di dati del primo grafico con un suo shift, mostrandone l'autocorrelazione. Effettivamente nel secondo grafico notiamo che il coefficiente di autocorrelazione assume valore 1 ad intervalli regolari, confermando l'ipotesi.

```python
################## Task5: Studio di correlazione ##################
from pandas.io.pytables import IndexCol
from ast import increment_lineno
import matplotlib.pyplot as mat
import numpy as np
import scipy.stats as sci
import pandas as pd

# salvo il csv in un dataframe, sistemiamo le colonne della tabella
di = pd.read_csv('Circolante_Basailicata.csv',usecols=[5,7],names=['Cilindrata','co2'])

# stampo csv
print(di)

# calcolo il coefficiente di correlazione r tra le due colonne e stampo
print(di.corr(method='pearson').iloc[0]['co2'])

# elevo alla seconda r
print(di.corr(method='pearson').iloc[0]['co2']**2)

group1 = di.Cilindrata
group2 = di['co2']
# converto liste in array
x = np.array(group1)
y = np.array(group2)

# calcolo pvalue
def f_test(group1, group2):
    f = np.var(group1, ddof=1)/np.var(group2, ddof=1)
    nun = x.size-1
    dun = y.size-1
    p_value = 1-sci.f.cdf(f, nun, dun)
    return f, p_value

f, pvalue = f_test(group1, group2)

# stampo p value
print(pvalue)
# true quindi è significativo
print(pvalue < 0.05)


plotter = mat.subplot()
plotter.plot(x, y, "yo")
```

```python
################## Task3: Who's Bigger? ##################

import networkx as nx
import wikipediaapi
import requests

#dataset delle persone ricercate
NAMES = [
            "Romolo",
            "Cesare",
            "Dante Alighieri",
            "Leonardo Da Vinci",
            "Benito Mussolini",
            "Umberto II di Savoia",
            "Rita Levi Montalcini",
            "Silvio Berlusconi",
            "Mario Draghi"
        ]

#dataset delle parole chiave da ricarcare
KEYWORDS = [
            "scoperta", "scoperto",
            "contribuito",
            "premiato", "premiata", "premio",
            "riconosciuto"
            ]

# Creazione un wikipedia api wrapper
wiki = wikipediaapi.Wikipedia(
        language='it',
        extract_format=wikipediaapi.ExtractFormat.WIKI
)

def getPageRank(name):

    # Creazione grafo vuoto
    G = nx.DiGraph()

    # Recupero pagina Wikipedia
    page = wiki.page(name)

    # Aggiunta nodi
    G.add_nodes_from([page.title])

    # Recupero link interni alla pagina
    for link in page.links.items():
        G.add_nodes_from([link[0]])
        G.add_edge(page.title, link[0])

    # Calcolo PageRank
    pr = nx.pagerank(G)

    # Estrazione PageRank
    for node, score in pr.items():
        if node == name:
            return node, score

def getPageViews(name):

    # Formattazione necessaria per la wikimedia API
    title = name.replace(" ", "_")

    # Costruzione url, recupera le visite alla pagina wikipedia effettuate da utenti reali, nel mese di marzo 2023
    url = f'https://wikimedia.org/api/rest_v1/metrics/pageviews/per-article/it.wikipedia.org/all-access/user/{title}/monthly/20230301/20230331'

    # Headers necessari per l'API
    headers = {'User-Agent': 'MediaWiki REST API docs examples/0.1 (https://www.mediawiki.org/wiki/API_talk:REST_API)'}

    # Recupero dei dati
    response = requests.get(url, headers=headers)
    data = response.json()

    # Estrazione numero di visite
    return data['items'][0]['views']

def getKeywordsCount(name):

    count = 0

    # Estrazione testo
    pagetext = wiki.page(name).text

    # Aggiornamento contatore
    for word in KEYWORDS:
        if word in pagetext.split():
            count += 1

    return count

def sortByPageRank(e):
    return e['pagerank']

def sortByPageViews(e):
    return e['pageviews']

def sortByTotalPoints(e):
    return e['rank_total']

candidates = []

for name in NAMES:
    node, score = getPageRank(name)
    views = getPageViews(name)
    keywords = getKeywordsCount(name)
    dic = dict(name=name, pagerank=score, pageviews=views, keywords=keywords)
    candidates.append(dic)


# Assegnazione punteggio per pageviews
# pageviews alto - punteggio alto
candidates.sort(reverse=True, key=sortByPageViews)
i = 0
for candidate in candidates:
    candidate['pageviewPoints'] = len(NAMES)-i
    i += 1


# Assegnazione punteggio per pagerank
# pagerank basso - punteggio alto
candidates.sort(key=sortByPageRank)
i = 0
for candidate in candidates:
    candidate['pagerankPoints'] = len(NAMES)-i
    i += 1


# Calcolo punteggio totale
# punti pagerank + punti pageviews + 2 punti per ogni keyword
for candidate in candidates:
    candidate['rank_total'] = candidate['pageviewPoints'] + candidate['pagerankPoints'] + (2*candidate['keywords'])

# print(candidates)

# Stampa della classifica
candidates.sort(reverse=True, key=sortByTotalPoints)
i = 1
print("Classifica Borda Ranking:")
for candidate in candidates:
    print(f"{i}) {candidate['name']}: {candidate['rank_total']} punti")
    i += 1
```

In questo task abbiamo ricercato quale tra diversi personaggi sia più grande secondo wikipedia, per determinare quale sia il più grande abbiamo definito delle parole chiave dove al ripetersi di esse veniva assegnato un punteggio.


Le metriche utilizzate per il ranking sono:
- pagerank: punteggio più alto (9) a chi ha il pagerank più basso
- pageviews: punteggio più alto (9) a chi ha il maggior numero di visualizzazioni sulla pagina wikipedia da parte di utenti umani
- parole chiave: vengono assegnati due punti ad ogni parola chiave trovata nel testo estratto dalla pagina wikipedia della persona interessata

```python
###############   Task1: T-test QI    ##################

import math
import statistics as st
from scipy.stats import skew
from scipy.stats import kurtosis
from scipy.stats import ttest_ind

### Dataset ###
qiF = [119, 126, 130, 121, 121, 117, 139, 130, 120, 114, 120]
qiM = [130, 113, 111, 117, 122, 118, 120, 114, 112, 116, 120]

print("Quozienti intelletivi:")
print(f"QI F: {qiF}")
print(f"QI M: {qiM}")

print(" ")

### Numerosità ###
n = len(qiF)
print(f"Numerosità dataset: {n}")

print(" ")

### Media ###
meanF = st.mean(qiF)
print(f"Media QI F: {meanF}")
meanM = st.mean(qiM)
print(f"Media QI M: {meanM}")

print(" ")

### Deviazione Standard ###
varF = sum((l-meanF)**2 for l in qiF) / (n-1)
devstdF = math.sqrt(varF)
print(f"Deviazione standard QI F: {devstdF}")
varM = sum((l-meanM)**2 for l in qiM) / (n-1)
devstdM = math.sqrt(varM)
print(f"Deviazione standard QI M: {devstdM}")

print(" ")

### Mediana ###
print(f"Mediana QI F: {st.median(qiF)}")
print(f"Mediana QI M: {st.median(qiM)}")

print(" ")

### Asimettria ###
asF = skew(qiF, bias=False)
print(f"Asimmetria F: {asF}")
asM = skew(qiM, bias=False)
print(f"Asimmetria M: {asM}")

print(" ")

### Curtosi ###
print(f"Curtosi F: {kurtosis(qiF, bias=False)}")
print(f"Curtosi M: {kurtosis(qiM, bias=False)}")

print(" ")

### Distribuzione Normale? ###
normF = "si" if (asF < 2 and asF > -2) else "no"
print(f"QI F distrib. normale? {normF}")
normM = "si" if (asM < 2 and asM > -2) else "no"
print(f"QI M distrib. normale? {normM}")

print(" ")
### Test ###
t_stat, p_value = ttest_ind(qiF, qiM)
print("Valore T: ", t_stat)
print("Valore P: ", p_value)

print(" ")

### Risultati ###
t1 = [95, 0.05]
res1 = "si" if (p_value < t1[1]) else "no"
print(f"Livello significatività {t1[0]}%: p < {t1[1]}, significativo? {res1}")
t2 = [99, 0.01]
res2 = "si" if (p_value < t2[1]) else "no"
print(f"Livello significatività {t2[0]}%: p < {t2[1]}, significativo? {res2}")
```

In questo task abbiamo verificato la significativita del T-test su un dataset, abbiamo effettuato diversi calcoli come la media, la deviazione standard, la mediana, l'assimetria, la curtosi che sono serviti per effettuare il calcolo delle significatività che poi intersecato con il valore di P (grado di libertà) possiamo determinare se sia effettivamente significativo o meno.

```python
###############   Task2: KS-test QI    ##################

import math
import statistics as st
from scipy.stats import norm

qi = [119, 126, 130, 121, 121, 117, 139, 130, 120, 114, 120]
qi.sort()

print(f"QI F: {qi}")

n = len(qi)
print(f"Numerosità dataset: {n}")

print(" ")

### calcolo media ###
mediaF = st.mean(qi)
print(f"Media QI F: {mediaF}")

print(" ")

### calcolo deviazione standard ###
varF = sum((l-mediaF)**2 for l in qi) / (n-1)
devstdF = math.sqrt(varF)
print(f"Deviazione standard QI F: {devstdF}")

print(" ")

### calcolo valore atteso ###
valAtteso = [i/n for i in range(n)]
print(f"Valore atteso QI F: {valAtteso}")

print(" ")

### calcolo vaolre reale tramite distribuzione normale ###
valReale = norm(mediaF, devstdF).cdf(qi)
print(f"Valore reale QI F: {valReale}")

print(" ")

### calcolo distanza ###
dist = [abs(r - a) for r, a in zip(valReale, valAtteso)]
print(f"Distanza valore reale - atteso: {dist}")

print(" ")

### calcolo distanza massima ###
maxDist = max(dist)
print(f"Distanza massima: {maxDist}")

print(" ")

### vaolre oracolo (da tabella) ###
valOracolo = 0.39122
print(f"Valore critico oracolo (n=11, alpha=0.05): {valOracolo}")

print(" ")

### stampa del risultato ###
res = "Distribuzione Normale" if (maxDist - valOracolo) else "Distribuzione non Normale"
print(f"Risultato KS-test su QI F: {res}")
```

in questo task abbiamo applicato il ks-test che confronta le funzioni cumulative di due distribuzioni e determina quanto le due distribuzioni siano simili, abbiamo come nel task precedente effettuato diversi calcoli per poter effettuare il calcolo di come sia la distribuzione in base al valore critico oracolo .

```python
###############   Task2: La curva ROC   ##################

import matplotlib.pyplot as plt
from sklearn import metrics

# Dati
negativeObserved = [34, 74, 100, 127, 153, 114, 19, 6, 4, 0]
positiveObserved = [2, 6, 10, 12, 23, 55, 75, 41, 30, 15]

# Calcolo valori cumulativi
negativeCumulative = [0]
positiveCumulative = [0]

for i in range(len(negativeObserved)):
	negativeCumulative.append(negativeObserved[i] + negativeCumulative[i])
	positiveCumulative.append(positiveObserved[i] + positiveCumulative[i])

# Calcolo FPR e TPR
FPR = []
TPR = []

for i in range(len(negativeObserved)):
	FPR.append(1 - negativeCumulative[i] / negativeCumulative[-1])
	TPR.append(1 - positiveCumulative[i] / positiveCumulative[-1])

# Calcolo AUC
AUC = []

for i in range(len(TPR)-1):
	AUC.append( (FPR[i]-FPR[i+1])*TPR[i] )

# Costruzione grafico
plt.scatter(FPR, TPR)
plt.title("ROC Curve")
plt.xlabel("False Positive Rate (FPR)")
plt.ylabel("True Positive Rate (TPR)")

# Annotazione AUC
plt.annotate('AUC: %.3f'%(sum(AUC)),
				xy=(0.95, 0.15), xycoords='data',
				ha='right', va="center",
				bbox=dict(boxstyle="round", fc="none", ec="gray"))

plt.show()
```

In questo task è stata utilizzata la curva ROC (Receiver-Operator Characteristic) per valutare un classificatore. La statistica più importante in questo caso è l'AUC (Area Under Curve), la quale rappresenta l'area sottesa al grafico della curva ROC; più il valore di AUC si avvicina ad 1, più il classificatore è valido. In questo caso abbiamo ottenuto AUC=0.894, concludiamo che il classificatore non è perfetto ma è comunque molto buono.

```python

###############   Task1: Regrassione lineare    ##################
import numpy as np
import matplotlib.pyplot as plt
import random as rn
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr

#Riprendendo l'algoritmo che genera altezza e peso dei cittadini in base alle statistiche che prese dall'Istat, i dati generati li utilizziamo come base per
#calcolare la retta di ragressione lineare

#Questo piccolo array contiene una media approssimativa di cittadini normopeso, sovrappeso e obeso su un campione di circa 1000 persone
indiciObesita = [493,361,115]

#Array contenente l'altezza dei cittadini
A = []
#Array contenente il peso dei cittadini
B = []


for i in range(indiciObesita[0]):
  raltezza = rn.uniform(1.50,2.10)
  bmi = rn.uniform(18.5,24.99)
  rpeso = bmi*(raltezza*raltezza)
  A.append(raltezza)
  B.append(rpeso)

# effettuo prima il controllo se i due dataset siano correlati tra loro attraverso il coefficiente di pearson, che ci indicherà se
# vi è una una correlazione lineare tra i tue dataset

p = pearsonr(A,B)
print("Il valore del coefficiente di pearson è: ",round(p[0],3))

# ora calcolo la regressione lineare
# alleno il modello
reg = LinearRegression()
A = np.reshape(A,[len(A),1])
B = np.reshape(B,[len(B),1])
reg.fit(A.reshape(-1, 1), B.reshape(-1, 1))

# calcolo: R^2, w1 e w0
rQ = reg.score(A.reshape(-1, 1), B.reshape(-1, 1))
w1= reg.coef_
w0= reg.intercept_
print(f"Questo è il valore di R^2: {round(rQ, 3)}")
print(f"Equazione: y = {round(w1[0][0], 3)}x {round(w0[0], 3)}")

print(" ")
#effettuo una predizione
pred = reg.predict([[3]])
print("Effettuo una predizione: se un uomo fosse alto 3 metri, avrebbe un peso di: ",round(pred[0][0], 1), "Kg")


# grafico risultato
plt.figure(figsize=(10, 5))
ax=plt.subplot()
ax.scatter(A, B)
ax.set_ylabel("Peso in Kg", fontsize="14")
ax.set_xlabel("Altezza in Cm", fontsize="14")
ax.set_title("Regressione lineare", fontsize="20")
plt.plot(A, reg.predict(A.reshape(-1, 1)), color='red')
plt.show()
```

In questo task abbiamo riutilizzato l'algoritmo per la generazione di peso e altezza per popolare il dataset, in questo caso è stato scelto implemetare solo la fascia dei normopeso perchè la correlazione tra peso e altezza è più forte, se ci fossero stati anche le altre fascie il valore di correlazione sarebbe stato inferiore.

Una volta calcolato il valore del coefficiente di correlazione di pearson e determinato che ci fosse una correlazione lineare, abbiamo calcolato la regressione lineare sui dataset.

Una volta calcolata abbiamo graficato il tutto, per determinare se e quanto fosse precisa abbiamo calcolato r^2 che ha un valore significativo per una predizione futura.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# Importiamo i dati
df = pd.read_csv('Social_Network_Ads.csv')

# Scegliamo feature e target
X = df[['Age','EstimatedSalary']]
y = df['Purchased']

# Creazione modello
model_without_Scaling = LogisticRegression(random_state=0, solver='lbfgs', multi_class='ovr').fit(X, y)
print(f'Accuracy senza scaling dei dati: {round(model_without_Scaling.score(X, y)*100,1)}%')

# Scaliamo i dati per una maggiore accuratezza
ss = StandardScaler()
X_scaled = ss.fit_transform(X)

# Divide il dataset in train set e test set
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.25, random_state=42)

# Creazione e addestramento del modello
reg = LogisticRegression(random_state=0, solver='lbfgs', multi_class='ovr')
reg.fit(X_train, y_train)

# y_pred = reg.predict(X_test)

# Valutazione del modello e stampa accuratezza
print(f'Accuracy con dati scalati: {round(reg.score(X_test, y_test)*100)}%')

# Creazione e stampa equazione
coef = reg.coef_[0]
intercept = reg.intercept_
eq = "y = "
for i,c in enumerate(coef):
	eq +=  "{:.2f}x_{} + ".format(c,i)
eq += "{:.2f}".format(intercept[0])
print(f"Equazione: {eq}")

# Costruzione grafico
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=plt.cm.Paired)
plt.title("Regressione Logistica")
plt.xlabel(f"Età (scalata, min: {X['Age'].min()} max: {X['Age'].max()})")
plt.ylabel(f"Salario (scalato, min: {X['EstimatedSalary'].min()} max: {X['EstimatedSalary'].max()})")

# Calcolo curva di decisione
x_min, x_max = X_train[:, 0].min() - .5, X_train[:, 0].max() + .5
y_min, y_max = X_train[:, 1].min() - .5, X_train[:, 1].max() + .5
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
Z = reg.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1]
Z = Z.reshape(xx.shape)

# Disegno curva di decisione
plt.contour(xx, yy, Z, levels=[0.5])
plt.show()
```

in questo task il modello utilizza come features l'età e il salario di alcuni utenti in modo da capire se essi siano propensi o meno ad acquistare un prodotto visto in una pubblicità presente sui social network.
Vale la pena notare come l'accuratezza sia salita di più del 20% scalando i dati, portando il modello da una scarsa accuratezza ad una molto migliore, a riconferma dell'importanza di fare un'analisi esplorativa dei dati, infatti i due parametri scelti presentavano valori molto distanti gli uni dagli altri.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB

# Importiamo e stampiamo i dati
df = pd.read_csv("email.csv")
print(df)

# Sostituzione stinghe nelle colonne con interi
# get_dummies() crea delle nuove colonne in modo da avere solo interi come valori
data = pd.get_dummies(df, drop_first=True)
# print(data)

# Suddivisione data in features e target
X = data.drop('valida_si', axis=1)
y = data['valida_si']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Creazione e fit del modello naive bayes
model = GaussianNB()
model.fit(X_train.values, y_train.values)

# Calcolo e stampa accuratezza
acc_train = model.score(X.values, y.values)
print("Accuracy modello = %0.3f " % acc_train)

# Utilizzo del modello per decidere se una nuova mail è spam o meno
predicted = model.predict([[1, 0, 0, 1, 1]])
res = "valida" if predicted[0] == 1 else "spam"
print(f"Predizione email: keyword=offerta, lunghezza=media, intestazione=si -> {res}")

# Controprova del modello simulando la precedente email ma senza intestazione
predicted = model.predict([[1, 0, 0, 1, 0]])
res = "valida" if predicted[0] == 1 else "spam"
print(f"Predizione email: keyword=offerta, lunghezza=media, intestazione=no -> {res}")
```

Il codice di cui sopra utilizza il modello naive bayes per determinare se una mail è spam o meno, esso utlizza: parole chiave, lunghezza della mail e presenza o meno di intestazione per effettuare la decisione. Il modello presenta un'accuratezza dell'80%, molto buona considerando i pochi dati per allenare il modello. Come esempio sono riportate due email, entrambe con keyword "offerta", entrambe di media lunghezza ma una con intestazione e l'altra senza: la prima email è valida, la seconda è classificata come spam.

```python
#Task2 lezione 14 Alberi di Decisione
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier

import pandas as pd
from sklearn.model_selection import train_test_split

# Importiamo e stampiamo i dati
df = pd.read_csv("email.csv")
print(df)

# Sostituzione stinghe nelle colonne con interi
# get_dummies() crea delle nuove colonne in modo da avere solo interi come valori
data = pd.get_dummies(df, drop_first=True)
# print(data)

# Suddivisione data in features e target
X = data.drop('valida_si', axis=1)
y = data['valida_si']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
clf = DecisionTreeClassifier()
clf.fit(X_train.values, y_train.values)

# Calcolo la precisione dell'algoritmo sul dataset
accuracy = clf.score(X_test.values, y_test.values)
print("Accuracy:", accuracy)

# Previsioni sui dati di test
predictions = clf.predict([[1, 0, 0, 1, 0]])

print(" ")

# Stampiamo delle previsioni
print("Effettuiamo delle previsioni")
res = "valida" if predictions[0] == 1 else "spam"
print(f"Predizione email: keyword=offerta, lunghezza=media, intestazione=si -> {res}")

res = "valida" if predictions[0] == 1 else "spam"
print(f"Predizione email: keyword=offerta, lunghezza=media, intestazione=no -> {res}")
```
